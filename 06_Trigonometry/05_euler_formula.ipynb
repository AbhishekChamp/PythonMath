```python
# e^ik = cos(k) + isin(k)
# me^ik = m(cos(k) + isin(k))

import numpy as np
import matplotlib.pyplot as plt
```

```python
k = np.pi / 4
m = 2.3

eul = m*np.exp(1j*k)
cis = m*(np.cos(k) + 1j*np.sin(k))

print(cis)
print(eul)
```

```python
mag = np.abs(eul)
ang = np.angle(eul)

print(mag, ang)

plt.polar([0, ang], [0, mag], 'r')
plt.polar(k, m, 'bo')

plt.show()
```

```python
def eulerFromCosSin():

    re = eval(input('cosine part: '))
    im = eval(input('sine part: '))

    print(re, im)

    m = np.sqrt( re**2 + im**2)
    k = np.arctan2(im, re)

    plt.polar([0, k], [0, m])
    plt.title('me$^{i\\phi}$, m=%g, $\\phi$=%g'%(m, k))
    plt.thetagrids([0, 45, 130, 200, 222.2])
    plt.show()
```
