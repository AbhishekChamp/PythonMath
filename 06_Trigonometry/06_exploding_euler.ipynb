```python
# me^ik
# Random m's and k's

# 40% pink lines
# 40% purple lines
# 20% green lines
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
nvects = 200
ms = np.random.rand(nvects)
ks = np.random.rand(nvects)*2*np.pi

for i in range(0, nvects):
    
    r = np.random.rand()

    if r < .4:
        clr = [1, .2, .7]
    elif r>.4 and r<.8:
        clr = [.7, .2, 1]
    else:
        clr = [0, 1, 0]

    plt.polar([0, ks[i]], [0, ms[i]], color=clr)

plt.axis('off')
plt.show()
```
