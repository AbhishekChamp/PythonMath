```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 8*np.pi, 1000)

r1 = np.random.rand()
r2 = np.random.rand()

x = np.cos(r1*t)
y = np.sin(r2*t)

plt.plot(x, y, 'k')
plt.axis('square')
plt.axis([-1.1, 1.1, -1.1, 1.1])
plt.axis('off')
plt.title('r$_1$=%s, r$_2$=%s'%(np.round(r1), np.round(r2)))

plt.show()
```

```python
t = np.linspace(0, 8*np.pi, 1000)

r1 = np.random.rand()
r2 = np.random.rand()

x = np.cos(r1*t)
y = np.sin(r2*t)

x = np.append(x, x[0])
y = np.append(y, y[0])

plt.plot(x, y, 'k')
plt.axis('square')
plt.axis([-1.1, 1.1, -1.1, 1.1])
plt.axis('off')
plt.title('r$_1$=%s, r$_2$=%s'%(np.round(r1), np.round(r2)))

plt.show()
```
