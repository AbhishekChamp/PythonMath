```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 3
b = 4

c = np.sqrt(a**2 + b**2)

# draw the triangle
plt.plot([0,a], [0, 0], 'k', linewidth=2)
plt.plot([0,0], [0,b], 'k')
plt.plot([0, a], [b, 0], 'k')

plt.plot([.3, .3], [0, .3], 'k', linewidth=1)
plt.plot([0, .3], [.3, .3], 'k')

plt.text(a/3, .25, 'a=' + str(a), fontsize=15)
plt.text(.1, b/3, 'b=' + str(b), fontsize=15)
plt.text(a/2+.1, b/2, 'c=' + str(c), fontsize=15)

plt.axis('square')
axlim = np.max((a, b)) + .5
plt.axis([-.5, axlim, -.5, axlim])
plt.axis('off')
plt.show()
```

```python
z = complex(3, 4)

mag1 = np.sqrt(np.real(z)**2 + np.imag(z)**2)
mag2 = np.abs(z)

print(mag1, mag2)
```

```python
ang1 = np.arctan2(np.imag(z), np.real(z))
ang2 = np.angle(z)

print(ang1, ang2)
```
