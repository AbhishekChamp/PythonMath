```python
# f(x) = 0, if x <= 0 && f(x) = -2x, if x > 0
```

```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math
```

```python
from sympy.abc import x

piece1 = 0
piece2 = -2*x
piece3 = x**3/10

fx = sym.Piecewise((piece1, x<0), (piece2, (x >=0) & (x < 3)), (piece3, x>=3))

fxx = sym.lambdify(x, fx)
xx = np.linspace(-3, 5, 1234)

plt.plot(xx, fxx(xx))
plt.show()
```

```python
# fx = x^3, for x <= 0 && log2(x), otherwise

piece1 = x**3
piece2 = sym.log(x,2)

fx = sym.Piecewise((piece1, x<= 0), (piece2, x > 0))

display(Math('f(x) = ' + sym.latex(fx)))

fxx = sym.lambdify(x, fx)
xx = np.linspace(-2, 2, 1000)

with plt.xkcd():
    plt.plot(xx, fxx(xx), 'k')
plt.xlim([-2, 2])
plt.ylim([-10, 3])

plt.show()
```
