```python
# d/dx cos(x) = -sin(x)
# d/dx -sin(x) = -cos(x)
# d/dx -cos(x) = sin(x)
# d/dx sin(x) = cos(x)
```

```python
import sympy as sym
from IPython.display import display, Math
```

```python
q = sym.symbols('q')

print(sym.diff( sym.cos(q) ))
print(sym.diff( sym.sin(q) ))
```

```python
f = sym.cos(q)

for i in range(0, 8):
    display(Math('\\frac{d}{dx}%s = %s' %(sym.latex(f), sym.latex(sym.diff(f)))))
    f = sym.diff(f)
```

```python
import sympy.plotting.plot as symplot

f = sym.cos(q)

for i in range(0, 4):

    if i==0:
        p = symplot(f, show=False, label =r'$%s$' %sym.latex(f))
    else:
        p.extend(symplot(f, show=False, label=r'$%s$' %sym.latex(f)))
    f = sym.diff(f)

p.legend = True
p.xlim = [-3, 3]
p.show()
```

```python
# fa(x) = cos(x + sin(x)) + a
# a = [0, 1, 2, 3]

x = sym.symbols('x')

for a in range(0, 4):
    fx = sym.cos(x + sym.sin(x)) + a
    if a == 0:
        p = symplot(fx, show=False, label =r'$%s$' %sym.latex(fx))
    else:
        p.extend(symplot(fx, show=False, label =r'$%s$' %sym.latex(fx)))

p.legend = True
p.show()
```

```python
x = sym.symbols('x')

for a in range(0, 4):
    fx = sym.diff(sym.cos(x + sym.sin(x)) + a)
    if a==0:
        p = symplot(fx, show=False, label=r'$%s$' %sym.latex(fx))
    else:
        p.extend(symplot(fx, show=False, label=r'$%s$' %sym.latex(fx)))

p.legend = True
p.show()
```
