```python
import sympy as sym
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display, Math
```

```python
x = sym.symbols('x')

fx = x**2
dfx = sym.diff(fx)
```

```python
# Leibniz notation
print("Leibniz Notation")
display(Math('f(x) = %s, \\quad \\frac{df}{dx}=%s' %(sym.latex(fx), sym.latex(dfx))))

# Lagrange notation
print("Lagrange Notation")
display(Math('f(x) = %s, \\quad f\'=%s' %(sym.latex(fx), sym.latex(dfx))))

# Newton notation
print("Newton Notation")
display(Math('f(x) = %s, \\quad \\dot{f}=%s' %(sym.latex(fx), sym.latex(dfx))))
```

```python
import sympy.plotting.plot as symplot

fx = 3 - x**3

p = symplot(fx, (x, -5, 5), show=False)
p.extend(symplot(sym.diff(fx), show=False))

p.xlim = [-3, 3]
p.ylim = [-10, 10]

p[1].line_color = 'r'
p[0].label = '$f(x) = %s$' %sym.latex(fx)
p[1].label = '$f(x) = %s$' %sym.latex(sym.diff(fx))

p.legend = True
p.show()
```

```python
# fx = 3 + 2x - 5x^2 + 7x^4
# gx = 4x^2 + x^5

# (f + g)' = f' + g'
# (f x g)' != f' x g'
# (f x g)' = f' x g + f x g'

fx = 3 + 2*x - 5 * x**2 + 7 * x**4
gx = 4*x**2 + x**5

dfx = sym.diff(fx)
dgx = sym.diff(gx)

d_fx_add_gx = sym.diff(fx + gx)
d_fx_prod_gx = sym.diff(fx * gx)

display(Math('f(x) = %s, \\quad f\'=%s' %(sym.latex(fx), sym.latex(dfx))))
display(Math('g(x) = %s, \\quad g\'=%s' %(sym.latex(fx), sym.latex(dgx))))

display(Math('(f + g)\' = %s' %sym.latex(d_fx_add_gx)))
display(Math('f\' + g\' = %s' %sym.latex((dfx + dgx))))

display(Math('(f \\times g)\' = %s' %sym.latex(sym.expand(d_fx_prod_gx))))
display(Math('f\' \\times g\' = %s' %sym.latex(sym.expand((dfx * dgx)))))
display(Math('f\' \\times g + f \\times g\' = %s' %sym.latex(sym.expand((dfx * gx + fx * dgx)))))
```
