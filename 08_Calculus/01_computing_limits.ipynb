```python
import sympy as sym
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display, Math
```

```python
x = sym.symbols('x')

fx = x**3

lim_pnt = 1.5
lim = sym.limit(fx, x, lim_pnt)

display(Math('\\lim_{x\\to %g} %s = %g' %(lim_pnt, sym.latex(fx), lim)))
```

```python
fxx = sym.lambdify(x, fx)
xx = np.linspace(-5, 5, 200)

plt.plot(xx, fxx(xx))
plt.plot(lim_pnt, lim, 'ro')

plt.show()
```

```python
fx = (x**2) / (x-2)
fxx = sym.lambdify(x, fx)
xx = np.linspace(1, 3, 102)

lim_pnt = 2

lim_1 = sym.limit(fx, x, lim_pnt, dir='+')
lim_2 = sym.limit(fx, x, lim_pnt, dir='-')

print(lim_1, lim_2)

plt.plot(xx, fxx(xx))
display(Math('\\lim_{x\\to %g^-} %s = %g' %(lim_pnt, sym.latex(fx), lim_2)))
```

```python
# f(x) = e^-x * sqrt(x+1)
# g(x) = cos(x + sin(x))

x = sym.symbols('x')

fx = sym.sqrt(x+1)*sym.exp(-x)
gx = sym.cos(x + sym.sin(x))

xx = np.linspace(0, 10, 100)

fxx = sym.lambdify(x, fx)
gxx = sym.lambdify(x, gx)

plt.plot(xx, fxx(xx), label='f(x)')
plt.plot(xx, gxx(xx), label='g(x)')

plt.legend()
plt.show()
```

```python
# compute the limits of f and g separately
lim_pnt = 5

lim_fx = sym.limit(fx, x, lim_pnt)
lim_gx = sym.limit(gx, x, lim_pnt)

# compute the limit of f/g
hx = fx/gx
lim_fgx = sym.limit(hx, x, lim_pnt)

display(Math('\\frac{\\lim_{x \\to %g} f(x)}{\\lim_{x \\to %g} g(x)} = \\frac{%g}{%g} = %g' %(lim_pnt, lim_pnt, lim_fx, lim_gx, lim_fx/lim_gx)))

display(Math('\\lim_{x\\to %g} \\frac{f(x)}{g(x)} = %g' %(lim_pnt, lim_fgx)))
```
