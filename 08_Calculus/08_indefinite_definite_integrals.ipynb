```python
import sympy as sym
from IPython.display import display, Math
```

```python
x = sym.symbols('x')

f = x

sym.integrate(f)
```

```python
sym.integrate(f, (x, 0, 1))
```

```python
p = sym.plotting.plot(f, show=False)
p.xlim = [0, 1]
p.ylim = [0, 1]
p.show()
```

```python
sym.integrate(f, (x, 0, 1))
```

```python
f = x**3 / (x-2)

intf = sym.integrate(f)

p = sym.plotting.plot(f, show=False)

p.extend(sym.plotting.plot(intf, (x, 2.1, 10), show=False, line_color='r'))

p[0].label = '$f(x) = %s$' %(sym.latex(f))
p[1].label = '$\\int f(x) dx = %s$' %(sym.latex(intf))

p.ylim = [-200, 200]
p.legend = True
p.show()
```

```python
# integral of f(x)' dx = f(x)
# f(x) = 2x^3 + sin(x)

x = sym.symbols('x')

fx = 2*x**3 + sym.sin(x)

diff_fx = sym.diff(fx)

integral_fx = sym.integrate(diff_fx)

display(Math('f(x) = %s' %sym.latex(fx)))
display(Math('f\' = %s' %sym.latex(diff_fx)))
display(Math('\\int (f\')dx = %s' %sym.latex(integral_fx)))
```
