```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
from IPython.display import display, Math
from mpl_toolkits.mplot3d import Axes3D
```

```python
row_vec = np.array([4, 2])
col_vec = np.array([ [-2], [3] ])

print(row_vec)
print(col_vec)
```

```python
display(Math(sym.latex(sym.simplify(row_vec))))
display(Math(sym.latex(sym.simplify(col_vec))))
```

```python
row_vec[0], row_vec[1]
```

```python
col_vec[0][0], col_vec[1][0]
```

```python
col_vec[0], col_vec[1]
```

```python
plt.plot([0, row_vec[0]], [0, row_vec[1]], 'r', label='row vector')
plt.plot([0, col_vec[0][0]], [0, col_vec[1][0]], 'b', label='column vector')

plt.axis('square')
plt.axis([-5, 5, -5, 5])
plt.grid()
plt.legend()
plt.show()
```

```python
v = np.array([3, 0, -4])
w = np.array([-1, 1, 3])

fig = plt.figure(figsize=plt.figaspect(1))
ax = fig.add_subplot(111, projection='3d')

ax.plot([0,v[0]], [0, v[1]], [0, v[2]], 'b', linewidth=3)
ax.plot([0, w[0]], [0, w[1]], [0, w[2]], 'r', linewidth=3)

ax.plot([-5, 5], [0, 0], [0, 0], '--', color=[.7, .7, .7])
ax.plot([0, 0], [-5, 5], [0, 0], '--', color=[.7, .7, .7])
ax.plot([-5, 5], [0, 0], [0, 0], '--', color=[.7, .7, .7])


ax.set_xlim3d(-5, 5)
ax.set_ylim3d(-5, 5)
ax.set_zlim3d(-5, 5)
plt.show()
```
