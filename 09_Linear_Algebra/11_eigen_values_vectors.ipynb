```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
M = np.random.randint(-5, 6, (5, 5))
M = M@M.T

eigvals, eigvecs = np.linalg.eig(M)

print(eigvals)

fig, ax = plt.subplots(1, 3)

ax[0].imshow(M)
ax[0].set_title('The matrix')

ax[1].imshow(np.diag(eigvals))
ax[1].set_title('Eigen Values')

ax[2].imshow(eigvecs)
ax[2].set_title('Eigen Vectors')

plt.show()
```

```python
Mv = M@eigvecs[:,0]
lv = eigvals[0]*eigvecs[:,0]

print(Mv)
print(lv)
```

```python
# Exercise 1
# 1. Extract eigenvectors and eigenvalues from M = [[1 2] [2 1]]
# 2. Plot lambda*v and Mv

M = np.array([[1, 2], [2, 1]])
d, V = np.linalg.eig(M)

lv = d[0]*V[:, 0]
Mv = M@V[:,0]

plt.plot([0, lv[0]], [0, lv[1]], 'b', label='$\\lambda v$', linewidth=4)
plt.plot([0, Mv[0]], [0, Mv[1]], 'r:', label='Mv', linewidth=4)

plt.axis('square')
plt.axis([-3, 3, -3, 3])

plt.legend()
plt.grid()
plt.show()
```

```python
M = np.array([[1, 2], [2, 1]])
d, V = np.linalg.eig(M)

lv = d[1]*V[:, 1]
Mv = M@V[:,1]

plt.plot([0, lv[0]], [0, lv[1]], 'b', label='$\\lambda v$', linewidth=4)
plt.plot([0, Mv[0]], [0, Mv[1]], 'r:', label='Mv', linewidth=4)

plt.axis('square')
plt.axis([-3, 3, -3, 3])

plt.legend()
plt.grid()
plt.show()
```

```python
# Exercise 2
# 1. Create a 6x6 symmetric random-numbers matrix
# 2. Extract eigenvectors
# 3. Compute dot products between all pairs of eigenvectors and show as a matrix

m = 6
Q = np.random.randn(m, m)
Q = Q@Q.T

d, V = np.linalg.eig(Q)

print(V)
```

```python
dps = np.zeros((m, m))

for i in range(m):
    for j in range(m):
        dps[i,j] = np.dot(V[:, i], V[:, j])

print(np.round(dps, 2))
```
