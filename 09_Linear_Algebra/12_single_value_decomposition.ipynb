```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
ein = Image.open('einstein.jpg')
print(np.shape(ein))

plt.imshow(ein)
plt.show()
```

```python
ein = np.mean(ein, 2)

print(np.shape(ein))

plt.imshow(ein)
plt.show()
```

```python
U, s, V = np.linalg.svd(ein)

print(np.shape(U))
print(np.shape(s))
print(np.shape(V))
```

```python
fig, ax = plt.subplots(1, 3)

ax[0].imshow(U)
ax[0].set_title('U')

ax[1].imshow(np.diag(s))
ax[1].set_title('$\\Sigma$')

ax[2].imshow(V)
ax[2].set_title('V')

plt.show()
```

```python
fig, ax = plt.subplots(1, 3)

ax[0].imshow(U)
ax[0].set_title('U')

ax[1].imshow(np.diag(np.log(s)))
ax[1].set_title('$\\Sigma$')

ax[2].imshow(V)
ax[2].set_title('V')

plt.show()
```

```python
plt.plot(s, 'ks-', markerfacecolor='w')
plt.show()
```

```python
plt.plot(s, 'ks-', markerfacecolor='w')
plt.xlim([-1, 50])
plt.title('Eigenspectrum of Einstein')
plt.show()
```

```python
# Exercise
# 1. Reconstruct the image from the SVD
# 2. Reconstruct Einstein but shuffle the singular values
# 3. Partially reconstruct Einstein using the first N components
```

```python
S = np.zeros(np.shape(ein))
for i in range(0, len(s)):
    S[i, i] = s[i]

rein = U@S@V

plt.subplot(1, 2, 1)
plt.imshow(ein)
plt.axis('off')
plt.title('original')

plt.subplot(1, 2, 2)
plt.imshow(rein)
plt.axis('off')
plt.title('SVD Version')

plt.show()
```

```python
# shuffle the order of the singular values
randorder = np.random.permutation(len(s))


S = np.zeros(np.shape(ein))
for i in range(0, len(s)):
    S[i, i] = s[randorder[i]]

rein = U@S@V

plt.subplot(1, 2, 1)
plt.imshow(ein)
plt.axis('off')
plt.title('original')

plt.subplot(1, 2, 2)
plt.imshow(rein)
plt.axis('off')
plt.title('SVD Version')

plt.show()
```
