```python
# xn+1 = rxn(1 - xn)
# x1 = .5
# n <= 500
# r = [1, 4] in 1000 steps

# For each value of r, plot the unique values in the last 10% of the vector x. r goes on the x-axis and x goes on the y-axis
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# %matplotlib notebook

N = 500

r = 1
x = .5+np.zeros(N)

for n in range(N-1):
    x[n+1] = r*x[n]*(1-x[n])

plt.plot(x, 's-')
plt.show()
```

```python
rs = np.linspace(1, 4, 1000)

N = 500
x = .5+np.zeros(N)

for ri in range(len(rs)):
    for n in range(N-1):
        x[n+1] = rs[ri]*x[n]*(1-x[n])

plt.plot(x, '-')
plt.show()
```

```python
rs = np.linspace(1, 4, 1000)

N = 500
x = .5+np.zeros(N)
endcap = np.arange(round(N*.9),N)

for ri in range(len(rs)):
    for n in range(N-1):
        x[n+1] = rs[ri]*x[n]*(1-x[n])

    u = np.unique(x[endcap])
    r = rs[ri]*np.ones(len(u))
    plt.plot(r, u, '.', markersize=1, color=[(np.sin(ri/len(rs)/2)+1)/2, 1-ri/len(rs),.5])

plt.show()
```

```python
rs = np.linspace(1, 4, 1000)

N = 500
x = .5+np.zeros(N)
endcap = np.arange(round(N*.9),N)

for ri in range(len(rs)):
    for n in range(N-1):
        x[n+1] = rs[ri]*x[n]*(1-x[n]) + .2*np.sin(2*np.pi*x[n])**2

    u = np.unique(x[endcap])
    r = rs[ri]*np.ones(len(u))
    plt.plot(r, u, '.', markersize=1, color=[(np.sin(ri/len(rs)/2)+1)/2, 1-ri/len(rs),.5])

plt.show()
```
