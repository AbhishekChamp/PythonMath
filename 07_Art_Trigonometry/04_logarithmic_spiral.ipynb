```python
# cos(t)
# e^-t
# x = cos(t)e^(t/k)
# y = sin(t)e^(t/k)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0, 10*np.pi, 1234)

k = -3

x = np.cos(t) * np.exp(t/k)
y = np.sin(t) * np.exp(t/k)

plt.plot(x, y, 'r')
plt.axis('square')
plt.axis('off')
plt.show()
```

```python
N = 150

ks = np.linspace(-6, -2, N)

for i in range(N):
    x = np.cos(t) * np.exp(t/ks[i])
    y = np.sin(t) * np.exp(t/ks[i])
    plt.plot(x, y, color=[i/N, i/N, 0])

plt.axis('square')
plt.axis('off')
plt.show()
```
