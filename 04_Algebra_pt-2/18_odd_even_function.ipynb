```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, 20)

fEven = x**2
fEvenNeg = (-x)**2

plt.plot(x, fEven)
plt.plot(x, fEvenNeg, 'ro')

plt.show()
```

```python
fOdd = x**3
fOddNeg = (-x)**3

plt.plot(x, fOdd)
plt.plot(x, fOddNeg, 'ro')
```

```python
thetha = np.linspace(-2*np.pi, 2*np.pi, 39)

cosfun = np.cos(thetha)
cosfunN = np.cos(thetha)

plt.plot(thetha, cosfun)
plt.plot(thetha, cosfunN, 'ro')
plt.title('cosine(x)')
plt.show()

sinfun = np.sin(thetha)
sinfunN = np.sin(-thetha)

plt.plot(thetha, sinfun)
plt.plot(thetha, sinfunN, 'ro')
plt.title('sine(x)')
plt.show()
```
