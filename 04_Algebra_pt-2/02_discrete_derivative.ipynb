```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1, 3, 4, -4]
```

```python
np.diff(x)
```

```python
v = np.arange(0, 11)
print(v)
print(np.diff(v))
print(np.diff(v, 2))
```

```python
x = np.linspace(-2, 2, 101)
f = x**2

df = np.diff(f)
dx = np.diff(x)

plt.plot(x, f, label='f')
plt.plot(x[:-1], df/dx, label='df/dx')

plt.axis([-2,2, -1, 2])
plt.grid()
plt.show()
```
