```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
z = complex(2, 3)

plt.plot(np.real(z), np.imag(z), 'ro')
```

```python
plt.plot(np.real(z), np.imag(z), 'ro')
plt.plot([0, np.real(z)], [0, np.imag(z)], 'r')

plt.xlabel('real')
plt.ylabel('imag')
plt.grid()
plt.axis([-4, 4, -4, 4])
```

```python
plt.plot(np.real(z), np.imag(z), 'ro')
plt.plot([0, np.real(z)], [0, np.imag(z)], 'r')
plt.plot([0, 0], [-4, 4], 'k')
plt.plot([-4, 4], [0, 0], 'k')

plt.xlabel('real')
plt.ylabel('imag')
plt.grid()
plt.axis([-4, 4, -4, 4])
```

```python
z1 = complex(1, -3)
z2 = complex(1, -1)
```

```python
z1_sum_z2 = z1+z2
z1_mul_z2 = z1*z2
```

```python
plt.plot([0, np.real(z1)], [0, np.imag(z1)], label='z1')
plt.plot([0, np.real(z2)], [0, np.imag(z2)], label='z2')
plt.plot([0, np.real(z1_sum_z2)], [0, np.imag(z1_sum_z2)], label='z1+z2')

plt.plot([0, 0], [-5, 5], 'k')
plt.plot([-5, 5], [0, 0], 'k')

plt.xlabel('Real')
plt.ylabel('Imag')
plt.grid()
plt.axis('square')
plt.axis([-5, 5, -5, 5])
plt.legend()
plt.show()
```
