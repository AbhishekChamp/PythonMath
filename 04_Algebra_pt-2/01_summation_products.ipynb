```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
l = [1, 3, 4, 1, 6]

print(sum(l))
print(np.sum(l))
```

```python
np.prod(l)
```

```python
np.cumsum(l)
```

```python
np.cumprod(l)
```

```python
plt.plot(l, 'rs-')
plt.plot(np.cumsum(l), 'bo-')

plt.legend(['set', 'cum'])
plt.show()
```

```python
a = [1, 2, 3, 4, 5]

sum_a = np.cumsum(a)
print(sum_a)

sum_a_sq = np.cumsum(np.square(a))
print(sum_a_sq)

prod_a = np.cumprod(a)
print(prod_a)

prod_a_sq = np.cumprod(np.square(a))
print(prod_a_sq)
```

```python
res_1 = ((sum_a / sum_a_sq) == (1/sum_a))
print(res_1)

res_2 = ((prod_a / prod_a_sq) == (1/ prod_a))
print(res_2)
```
