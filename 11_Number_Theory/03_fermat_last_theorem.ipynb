```python
# x^n + y^n = z^n
# Has no integer solutions for n > 2
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ints = np.arange(1, 101)

Z = np.zeros((len(ints), len(ints)))

for x in ints:
    for y in ints:
        Z[x-1, y-1] = (x**3 + y**3)**(1/3)
```

```python
zInts = Z%1 == 0
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].imshow(Z)
ax[0].set_xlabel('y')
ax[0].set_ylabel('x')
ax[0].set_title('$z = \sqrt[3]{x^3+y^3}$', fontsize=16)

ax[1].imshow(zInts)
ax[1].set_xlabel('y')
ax[1].set_ylabel('x')
ax[1].set_title('Whether $z$ is an int', fontsize=16)

plt.show()
```

```python
# Reproduce the Boolean map for powers of 2
```

```python
ints = np.arange(1, 101)

Z = np.zeros((len(ints), len(ints)))

for x in ints:
    for y in ints:
        Z[x-1, y-1] = (x**2 + y**2)**(1/2)
```

```python
zInts = Z%1 == 0
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))

ax[0].imshow(Z)
ax[0].set_xlabel('y')
ax[0].set_ylabel('x')
ax[0].set_title('$z = \sqrt{x^2+y^2}$', fontsize=16)

ax[1].imshow(zInts)
ax[1].set_xlabel('y')
ax[1].set_ylabel('x')
ax[1].set_title('Whether $z$ is an int', fontsize=16)

plt.show()
```
