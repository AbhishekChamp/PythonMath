```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
s1 = np.array([])
s2 = np.array([])
s3 = np.array([])
s4 = np.array([])

xx = np.arange(1, 81)

for a in xx:
    s1 = np.append(s1, a / (a+1))
    s2 = np.append(s2, (a + 1) / a)
    s3 = np.append(s3, a / sum(np.arange(1, a+1)))
    s4 = np.append(s4, (-1)**a / np.sqrt(a))
```

```python
fig, ax = plt.subplots(1, figsize=(10, 7))
plt.plot(xx, s1, label='$a / (a+1)$')
plt.plot(xx, s2, label='$(a+1) / a$')
plt.plot(xx, s3, label='$a/\Sigma a$')
plt.plot(xx, s4, label='$-1^a/\sqrt{a}$')
plt.legend(loc='upper right', fontsize=16)

plt.show()
```

```python
# Generate the same sequences via "vectorization" instead of in a for-loop
```

```python
s1 = np.array([])
s2 = np.array([])
s3 = np.array([])
s4 = np.array([])

xx = np.arange(1, 81)

s1 = xx / (xx+1)
s2 = (xx + 1) / xx
s3 = xx / np.cumsum(xx)
s4 = (-1)**xx / np.sqrt(xx)
```

```python
fig, ax = plt.subplots(1, figsize=(10, 7))
plt.plot(xx, s1, label='$a / (a+1)$')
plt.plot(xx, s2, label='$(a+1) / a$')
plt.plot(xx, s3, label='$a/\Sigma a$')
plt.plot(xx, s4, label='$-1^a/\sqrt{a}$')
plt.legend(loc='upper right', fontsize=16)

plt.show()
```
