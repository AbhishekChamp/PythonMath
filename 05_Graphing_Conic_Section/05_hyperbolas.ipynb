```python
# ( (y-k)^2 / b^2 ) - ( (x-h)^2 / a^2 ) - 1 = 0
# Center: (h, k)
# Radii: a, b
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1
b = .5
h = 1
k = 2

axlim = 2*np.max((a,b)) + np.max((abs(h), abs(k)))
x = np.linspace(-axlim, axlim, 100)
X, Y = np.meshgrid(x, x)

# Create the function
hyperbola = (X-h)**2/a**2 - (Y-k)**2/b**2 - 1

plt.contour(X, Y, hyperbola, 0)
plt.plot(h, k, 'go')

plt.plot([-axlim, axlim], [0, 0], 'k--')
plt.plot([0, 0], [-axlim, axlim], 'r--')

plt.gca().set_aspect('equal')
```

```python
n = 16
ab = np.linspace(1, 5, n)

axlim = 8
x = np.linspace(-axlim, axlim, 100)
X, Y = np.meshgrid(x, x)

for i in range(0, n):
    F1 = X**2/ab[i]**2 - Y**2/ab[i]**2 - 1
    plt.contour(F1, 0, colors=[i/n, 0, i/n])

    F2 = -X**2/ab[i]**2 + Y**2/ab[i]**2 - 1
    plt.contour(F2, 0, colors=[0, i/n, i/n])

plt.axis('off')
plt.gca().set_aspect('equal')
plt.show()
```
