```python
# 4p(y-k) = (x-h)^2
# y = ( ( 1 / 4 ) * (x-h)^2 ) + k

import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1 # 1/4p
h = 1
k = -2
n = 100

# x-axis point to evaluate the function
x = np.linspace(h-5, h+5, n)

# create the function
y = a*(x-h)**2 + k

plt.plot(x, y)
plt.show()
```

```python
a = 1 #1/p
h = 1
k = -2
n = 100

# x-axis point to evaluate the function
x = np.linspace(h-5, h+5, n)

# create the function
y = a*(x-h)**2 + k

plt.plot(y, x)
plt.show()
```

```python
a = 2
p = 1 / (4*a)
h = 2
k = 1
n = 100

# x-axis point to evaluate the function
x = np.linspace(h-1, h+1, n)

# create the function
y = a*(x-h)**2 + k


# Plotting parabola
plt.plot(x, y, label='parabola')

# Plot the vertex
plt.plot(h, k, 'ro', label='vertex')

# Plot the focus
plt.plot(h, k+p, 'go', label='focus') 

# Plot the directrix
d = k-p
plt.plot(x[[0, -1]], [d, d], label='directrix')

plt.grid()
plt.legend()
plt.show()
```
