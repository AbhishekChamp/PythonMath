```python
import numpy  as np
import matplotlib.pyplot as plt
```

```python
X, Y = np.meshgrid(range(0, 10), range(0, 15))
X, Y
```

```python
plt.subplot(121)
plt.pcolormesh(X, edgecolor='k', linewidth=.1)

plt.subplot(122)
plt.pcolormesh(Y, edgecolor='k', linewidth=.1)

plt.show()
```

```python
Fxy = np.cos(X) + np.sin(Y)

plt.imshow(Fxy)
plt.show()
```

```python
x = np.linspace(-np.pi, 2*np.pi, 40)
y = np.linspace(0, 4*np.pi, 72)

X, Y = np.meshgrid(x, y)

Fxy = np.cos(X) + np.sin(Y)

plt.imshow(Fxy)
# plt.pcolormesh(Fxy, edgeColor='k', linewidth=.1)  # Use only for low resolution
plt.show()
```

```python
# 2D Gaussian / Image Smoothing Kernel / Blur
# Z = e^ ( -( X^2 + Y^2) / s )

x = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, x)
s = 2

gauss2d = np.exp( -(X**2 + Y**2) / s )

plt.imshow(gauss2d)
plt.axis('off')
plt.show()
```

```python
# 2D Gaussian / Image Smoothing Kernel / Blur

x = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, x)
s = .5

gauss2d = np.exp( -(X**2 + Y**2) / s )

plt.imshow(gauss2d)
plt.axis('off')
plt.show()
```
