```python
A = np.zeros((10, 14))
print(np.shape(A))
```

```python
A = np.zeros((10, 14))

for ri in range(0, np.shape(A)[0]):
    for ci in range(0, np.shape(A)[1]):
        A[ri, ci] = 3*ri - 4*ci

print(A)
plt.imshow(A)
plt.plot([0, 3], [8, 2], 'r', linewidth=4)

plt.show()
```

```python
A = np.zeros((10, 14))

for ri in range(0, np.shape(A)[0]):
    for ci in range(0, np.shape(A)[1]):
        A[ri, ci] = 3*ri - 4*ci

plt.imshow(A)
plt.plot([0, 3], [8, 2], 'r', linewidth=4)

for i in range(0, np.shape(A)[0]):
    for j in range(0, np.shape(A)[1]):
        plt.text(j, i, A[i, j], horizontalalignment='center', verticalalignment='center')

plt.set_cmap('Purples')

plt.show()
```

```python
N = 10

C = np.zeros((N,N))

for i in range(0, N):
    for j in range(0, N):
        C[i,j] = (-1)**(i+j)

plt.imshow(C)
plt.set_cmap('gray')
plt.tick_params(labelleft=False, labelbottom=False)
plt.show()
```
